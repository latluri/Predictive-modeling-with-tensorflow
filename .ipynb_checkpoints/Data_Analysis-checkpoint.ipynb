# Data Analysis phase
This script analyses the "Movie_metadata.csv" file and studies the data with the help of visualizations

```python
import time
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Code to read the data
data = pd.read_csv('C:\\Users\\Pavan\\Desktop\\rang\\movie_rating_prediction-master\\movie_metadata.csv', sep=',',header=0,encoding='latin-1')
```

```python
#Code to generate box plots of all the numeric columns
for column in data.select_dtypes(exclude=['object']):
        plt.figure().suptitle(column, fontsize=14, fontweight='bold')
        data.boxplot(column)

plt.show()
```

From box plots, it is clear that there is a category of movies that is highly influencing the plots
For example, consider the facebook likes, most of these columns are filled with zeros due to which a even 2000 would seem like an outlier. However, they are not supposed to be considered in that way.
Binning these columns would be a good approach



```python
#Histogram of imdbscores
data.imdb_score.hist()
```

```python
#Binning the imdb scores
data.imdb_score=pd.cut(data['imdb_score'],[0,4.0,6.0,8.0,10],labels=False)

#Histogram of imdbscores binned
data.imdb_score.hist()
```

(Left to right) worst movies,Average movies,Good movies and top Movies

```python
#Histogram of movie_facebook_likes
data.movie_facebook_likes.hist()
```

```python
#Binning
data.movie_facebook_likes=pd.cut(data['movie_facebook_likes'],[182.0, 613.8, 1000.0, 14000.0, 27200.0, 349000.0],labels=False)
```

```python
#Histogram of movie_facebook_likes binned
data.movie_facebook_likes.hist()
```

facebook likes are binned into five categories
The values from left to right indicate lowest to highest categories. It may so happen that each of this category has its own impact on imdb score and it may not be linear.
For example, movies with either lowest or highest likes may have correlation with imdb scores.Hence this binning could be helpful.

```python
#Pivot table to get top actors with gross
pivot=data.pivot_table(index=["actor_1_name"], values=["gross"], aggfunc=np.mean).sort_values(['gross'],ascending=False).head(10)
pivot.plot(kind="bar",color="b",label=pivot)
plt.show()
```

Top Actors and their mean gross collections

```python
#missing values in numeric columns
new={}
for column in data.columns:
    new[column]=pd.value_counts(pd.isna(data[column]).values, ascending=True,bins=2).values[0]

pd.Series(new).plot(kind="bar",title="Missing values")
plt.show()
```
